```python
import pandas as pd
import numpy as np
```

### 데이터 읽기 (Kaggle)

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)

# df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/lol.csv',sep='\t')
df.head()
```

### 결측치 제거

```python
df['channelId'] = np.nan
df.dropna(axis='columns', how='all')
```

### 데이터 수정

```python
df.drop(df['like rate'], axis=0)
```

```python
# if 'like rate' in df.columns():
#     df.drop(df['like rate'])

df['like rate'] = (df['likes']/(df['likes'] + df['dislikes']) * 100).round(1)
df
```

```python
df['recommendation'] = 'No'
df
```

```python
like_filter = df['like rate'] > 99
df.loc[like_filter, 'recommendation'] = 'Yes'
df
```

```python
df[df['recommendation']=='Yes'].shape[0]
```

### Row 삭제

```python
### view_count 가 1만회 미만일 경우 df 에서 제거
filt = df['view_count'] < 10000
df[filt]

df.drop(index=df[filt].index)
```

```python
df.loc[60400] = ['유튜브 제목', '채널 이름', 10, 0, 0, 0, 0, np.nan, '2024-08-07', np.nan, 'No']
df
```

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]]
df
```

## 13. 함수 적용

```python
def add_percent(num):
    if pd.notnull(num):
        return str(num)+"%"
    return num

df['like rate'] = df['like rate'].apply(add_percent)
df
```

## 퀴즈

```python
import pandas as pd
data = {
    '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
    '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
df
```

```python
### 전체 데이터 중에서 영화 정보만 출력
df['영화']
```

```python
### 영화 평점 정보만 출력
df[['영화', '평점']]
```

```python
### 추천 점수: 관객수 * 평점 // 100

df['추천점수'] = df['관객 수']*df['평점']//100
df
```

```python
### 개봉년도 내림차순
df.sort_values(by='개봉 연도', ascending=False)
```
